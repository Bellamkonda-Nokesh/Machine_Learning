```python
#Heeatmaps
import pandas as pd
import seaborn as sns
data = {'Age': [22, 25, 29, 24, 23, 27, 30],
'Salary': [40000, 50000, 60000, 55000, 45000, 70000, 80000],
'Experience': [1, 3, 5, 2, 4, 6, 7]}
df = pd.DataFrame(data)
sns.heatmap(df.corr(numeric_only = True),annot=True,cmap="Reds") #Blues,BuPu,Greens,Reds,Y1OrBr
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data (replace with your actual data)
data = {
    'Math': {'Male': 22, 'Female': 41},
    'Science': {'Male': 10, 'Female': 25},
    'History': {'Male': 13, 'Female': 35},
    'Computer': {'Male': 27, 'Female': 55},
    'English': {'Male': 21, 'Female': 40}
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
subjects = df.columns
male_counts = df.loc['Male']
female_counts = df.loc['Female']

width = 0.35  # Width of the bars

fig, ax = plt.subplots(figsize=(10, 6))  

male_bars = ax.bar(subjects, male_counts, width, label='Male', color='darkblue')
female_bars = ax.bar(subjects, female_counts, width, bottom=male_counts, label='Female', color='orange')

# Add labels, title, and legend
ax.set_xlabel('Subject')
ax.set_ylabel('Number of Students')
ax.set_title('Number of Students by Subject and Gender')
ax.legend()

# Improve layout
plt.xticks(rotation=45, ha='right')  
plt.tight_layout() 

# Show the plot
plt.show()
```
