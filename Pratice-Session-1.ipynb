```python
df=pd.read_csv("Football_dataset.csv")
df
```

```python
Team= df['Team'].unique()
Team
```

```python
Teams_participated= df['Team'].nunique()
print(f"the number of teams participated in the tournament:{Teams_participated}")
```

```python
df['Discipline'] = df['Red Cards'] + df['Yellow Cards']
top_teams = df.groupby('Team')['Discipline'].sum().nlargest(2)
print("Top two teams with highest discipline:")
print(top_teams)
```

```python
df['Discipline'] = df['Red Cards'] + df['Yellow Cards']
top_teams = df.groupby('Team')['Discipline'].sum().nsmallest(2)
print("Top two teams with highest discipline:")
print(top_teams)
```

```python
average_yellow=df.groupby("Team")['Yellow Cards'].mean()
print(average_yellow)
```

```python
average_yellow=df['Yellow Cards'].mean()
print(average_yellow)
```

```python
team_goals=df[df['Goals Scored']>5]
num_team_more_5_goals=team_goals['Team'].nunique()
print(f"{num_team_more_5_goals} team scored more then 5 goals")
print("\n the teams the scored more then 5 goals ")
print(team_goals[['Team','Goals Scored']])
```
